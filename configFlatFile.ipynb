##### Fig mã hoá chữ cái

```python
columns_encoding = ['city', 'name', 'officialName']

def decode_unicode_escape(text):
    if pd.isna(text):
        return text
    try:
        return text.encode('utf-8').decode('unicode_escape')
    except Exception:
        return text
    
for col in columns_encoding:
    if col in team_df.columns:
        team_df[col] = team_df[col].apply(decode_unicode_escape)

print(team_df[columns_encoding].head(10))
```

##### Lược bỏ các alpha2code và alpha3code chỉ giữ ID cho phù hợp với chuẩn hoá của SQL

```python
def clean_df(area_str):
    try:
        area_dict = ast.literal_eval(area_str)
        area_id = int(area_dict.get("id", 0))
        alpha3 = area_dict.get("alpha3code", "")
        
        if alpha3 =="XEN" or area_id == 0:
            return 826
        
        return area_id
    except:
        return None
    
team_df["area"] = team_df["area"].apply(clean_df)

output_path = os.path.join(path_, "teams_config.csv")
team_df.to_csv(output_path, index=False, encoding="utf-8-sig")
print(f"File auto save!!!: {output_path}")
```

#### TABLE MATCH * (England, European, France, Germany, Italy, Spain, World_Cup)

```python


def nomalize_matchTeams(df):
    team_rows = []
    for _, row in df.iterrows():
        match_id = row['wyId']
        teams_dict = row['teamsData']

        if isinstance(teams_dict, str):
            try:
                teams_dict = ast.literal_eval(teams_dict)
            except:
                continue

        for team_id_str, team_info in teams_dict.items():
            team_id = team_info.get('teamId')
            side = team_info.get('side')
            coach_id = team_info.get('coachId', None)
            team_rows.append({
                'matchID': match_id,
                'teamID': team_id,
                'side': side,
                'coachID': coach_id
            })
    return pd.DataFrame(team_rows)
```

#### TABLE MATCHREFEREES

```python
def normalize_match_referees(df):
    referees_rows =[]

    for _, row in df.iterrows():
        match_id = row.get('wyId') or row.get('matchId')
        referees_data = row.get('referees', None)

        if referees_data is None or (isinstance(referees_data, float) and pd.isna(referees_data)):
            continue

        try:
            if isinstance(referees_data, str):
                if referees_data.strip() in ['', '[]', 'None']:
                    continue
                referees = ast.literal_eval(referees_data)
            else:
                referees = referees_data
        except Exception:
            continue
        
        if not isinstance(referees, (list, tuple)) or len(referees) == 0:
            continue

        for ref in referees:
            if not isinstance(ref, dict):
                continue


            referee_id = ref.get('refereeId') or ref.get('id')
            role = ref.get('role') or ref.get('Rrole')

            if referee_id is None or role is None:
                continue
            
            referees_rows.append({
                'matchID': match_id,
                'refereeID': referee_id,
                'Rrole': role
            })

    return pd.DataFrame(referees_rows)
```

##### Table Referees

```python
def extract_referees(df, offset = 1000000):
    ref_rows = []

    for _, row in df.iterrows():
        referee_id = row.get("wyId")
        Sname = row.get("shortName")
        Fname = row.get("firstName")
        Mname = row.get("middleName")
        Lname = row.get("lastName")
        birthDate = row.get("birthDate")

        person_id = referee_id + offset if referee_id is not None else None

        ref_rows.append({
            "refereeID": referee_id,
            "personID": person_id,
            "Fname": Fname,
            "Mname": Mname,
            "Lname": Lname,
            "Sname": Sname,
            "birthDate": birthDate
        })

    return pd.DataFrame(ref_rows)
```

```python

def extract_unique_areas(df, cols=['passportArea', 'birthArea']):
    area_rows = []

    for col in cols:
        for area_str in df[col]:
            if isinstance(area_str, str):
                try:
                    area_dict = ast.literal_eval(area_str)
                    area_id = int(area_dict.get('id', 0))
                    name_area = area_dict.get('name', '')
                    alpha2 = area_dict.get('alpha2code', '')
                    alpha3 = area_dict.get('alpha3code', '')

                    area_rows.append({
                        'areaID': area_id,
                        'nameArea': name_area,
                        'alpha2code': alpha2,
                        'alpha3code': alpha3
                    })
                except:
                    continue

    # Chuyển sang DataFrame và loại bỏ trùng lặp theo areaID
    df_areas = pd.DataFrame(area_rows).drop_duplicates(subset='areaID').reset_index(drop=True)
    return df_areas
```

```python
offset_map = {
    "England" : 0,
    "European_Championship": 700000,
    "France": 800000,
    "Germany": 1500000,
    "Italy": 2100000,
    "Spain": 2800000,
    "World_Cup": 3500000
}
```

```python
def parse_tags(tag_str):
    if pd.isna(tag_str):
        return []

    # Nếu input đã là list
    if isinstance(tag_str, list):
        # bóc id nếu là dict
        return [t["id"] if isinstance(t, dict) else t for t in tag_str]

    try:
        tags = ast.literal_eval(tag_str)

        # tags có thể là list các dict → bóc id
        return [t["id"] if isinstance(t, dict) else t for t in tags]
    except:
        return []
```

```python
def split_events_tags(df, league_name):
    offset = offset_map.get(league_name, 0)

    df['tags_parsed'] = df['tags'].apply(parse_tags)
    df["eventRecordID"] = df.index + 1 + offset

    events_df =df.drop(columns=["eventId", "tags", "subEventName", "positions", "eventName", "id"], errors="ignore")

    eventtag_rows = []

    for _, row in df.iterrows():
        event_id = row["eventRecordID"]
        for tag_id in row["tags_parsed"]:
                eventtag_rows.append({
                    "eventRecordID": event_id,
                    "tagID": tag_id
                })
    
    eventtags_df = pd.DataFrame(eventtag_rows)
    
    return events_df, eventtags_df
```
